```python
#!/usr/bin/env python
# vim:fileencoding=utf-8
import sys
import scipy.signal as sg
import matplotlib.pyplot as plt
import soundfile as sf
import matplotlib

#可視化ライブラリ
import seaborn as sns
```

```python
#音楽ファイル
NHKRadio_file = './Input/01_Radio/NHKRadio.wav'
NHKRadio_whitenoise_file = './Input/01_Radio/NHKRadio_whitenoise.wav'
NHKBusiness_file = './Input/02_Business/NHKBusiness.wav'
NHKBusiness_whitenoise_file = './Input/02_Business/NHKBusiness_whitenoise.wav'
Classic_file = './Input/03_Classic/Classic.wav'
Classic_whitenoise_file = './Input/03_Classic/Classic_whitenoise.wav'
```

```python
plt.close("all")

# wavファイル読み込み
NHKRadio_wav, NHKRadio_fs = sf.read(NHKRadio_file)

# ステレオ2chの場合、モノラル音源に変換(左右の各音を2で割った音を足して作成．)
if(NHKRadio_wav.shape[1] == 1):
    NHKRadio_wavdata = NHKRadio_wav
    print(NHKRadio_wav.shape[1])
else:
    NHKRadio_wavdata = (0.5 * NHKRadio_wav[:, 1]) + (0.5 * NHKRadio_wav[:, 0])
```

```python
# wavファイル読み込み
NHKBusiness_wav, NHKBusiness_fs = sf.read(NHKBusiness_file)

# ステレオ2chの場合、モノラル音源に変換(左右の各音を2で割った音を足して作成．)
if(NHKBusiness_wav.shape[1] == 1):
    NHKBusiness_wavdata = NHKBusiness_wav
    print(NHKBusiness_wav.shape[1])
else:
    NHKBusiness_wavdata = (0.5 * NHKBusiness_wav[:, 1]) + (0.5 * NHKBusiness_wav[:, 0])
```

```python
# wavファイル読み込み
NHKBusiness_whitenoise_wav, NHKBusiness_whitenoise_fs = sf.read(NHKBusiness_whitenoise_file)

# ステレオ2chの場合、モノラル音源に変換(左右の各音を2で割った音を足して作成．)
if(NHKBusiness_whitenoise_wav.shape[1] == 1):
    NHKBusiness_whitenoise_wavdata = NHKBusiness_whitenoise_wav
    print(NHKBusiness_whitenoise_wav.shape[1])
else:
    NHKBusiness_whitenoise_wavdata = (0.5 * NHKBusiness_whitenoise_wav[:, 1]) + (0.5 * NHKBusiness_whitenoise_wav[:, 0])
```

```python
# wavファイル読み込み
Classic_wav, Classic_fs = sf.read(Classic_file)

# ステレオ2chの場合、モノラル音源に変換(左右の各音を2で割った音を足して作成．)
if(Classic_wav.shape[1] == 1):
    Classic_wavdata = Classic_wav
    print(Classic_wav.shape[1])
else:
    Classic_wavdata = (0.5 * Classic_wav[:, 1]) + (0.5 * Classic_wav[:, 0])
```

```python
x_label = "Histgram of NHK Radio English Signal"
y_label = ""
sns.utils.axlabel(xlabel=x_label, ylabel=y_label, fontsize=16)
sns.distplot(NHKRadio_wavdata,bins=1000, kde=False).set(yscale='log')
plt.savefig('./Output/SoundFile_Signal/Signal_NHKRadio_Histgram.png')
plt.show()
```
