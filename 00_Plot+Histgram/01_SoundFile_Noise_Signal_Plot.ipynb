```python
#!/usr/bin/env python
# vim:fileencoding=utf-8
import sys
import scipy.signal as sg
import matplotlib.pyplot as plt
import soundfile as sf
import matplotlib
```

```python
#音楽ファイル
whitenoise_file = './Input/04_Noise/whitenoise_NHKRadio.wav'
pinknoise_file = './Input/04_Noise/pinknoise_NHKRadio.wav'
brownnoise_file = './Input/04_Noise/brownnoise_NHKRadio.wav'
tpdfnoise_file = './Input/04_Noise/tpdfnoise_NHKRadio.wav'
```

```python
plt.close("all")

# wavファイル読み込み
whitenoise_wav, whitenoise_fs = sf.read(whitenoise_file)

# ステレオ2chの場合、モノラル音源に変換(左右の各音を2で割った音を足して作成．)
if(whitenoise_wav.shape[1] == 1):
    whitenoise_wavdata = whitenoise_wav
    print(whitenoise_wav.shape[1])
else:
    whitenoise_wavdata = (0.5 * whitenoise_wav[:, 1]) + (0.5 * whitenoise_wav[:, 0])
```

```python
# wavファイル読み込み
pinknoise_wav, pinknoise_fs = sf.read(pinknoise_file)

# ステレオ2chの場合、モノラル音源に変換(左右の各音を2で割った音を足して作成．)
if(pinknoise_wav.shape[1] == 1):
    pinknoise_wavdata = pinknoise_wav
    print(pinknoise_wav.shape[1])
else:
    pinknoise_wavdata = (0.5 * pinknoise_wav[:, 1]) + (0.5 * pinknoise_wav[:, 0])
```

```python
# wavファイル読み込み
brownnoise_wav, brownnoise_fs = sf.read(brownnoise_file)

# ステレオ2chの場合、モノラル音源に変換(左右の各音を2で割った音を足して作成．)
if(brownnoise_wav.shape[1] == 1):
    brownnoise_wavdata = brownnoise_wav
    print(brownnoise_wav.shape[1])
else:
    brownnoise_wavdata = (0.5 * brownnoise_wav[:, 1]) + (0.5 * brownnoise_wav[:, 0])
```

```python
# wavファイル読み込み
tpdfnoise_wav, tpdfnoise_fs = sf.read(tpdfnoise_file)

# ステレオ2chの場合、モノラル音源に変換(左右の各音を2で割った音を足して作成．)
if(tpdfnoise_wav.shape[1] == 1):
    tpdfnoise_wavdata = tpdfnoise_wav
    print(tpdfnoise_wav.shape[1])
else:
    tpdfnoise_wavdata = (0.5 * tpdfnoise_wav[:, 1]) + (0.5 * tpdfnoise_wav[:, 0])
```

```python
# 冒頭から30秒分プロット
#fig = plt.figure(1)
#,dpi=200
fig = plt.figure(figsize=(16,8))
matplotlib.rcParams.update({'font.size': 18})

ax = fig.add_subplot(411)
ax.plot(whitenoise_wavdata[:whitenoise_fs*30])
ax.set_title("Whitenoise based on NHK Radio English Signal")
ax.set_xlabel("time [pt]")
ax.set_ylabel("amplitude")
ax.set_ylim([-0.04, 0.04])

ax = fig.add_subplot(412)
ax.plot(pinknoise_wavdata[:pinknoise_fs*30])
ax.set_title("Pinknoise based on NHK Radio English Signal")
ax.set_xlabel("time [pt]")
ax.set_ylabel("amplitude")
ax.set_ylim([-0.04, 0.04])

ax = fig.add_subplot(413)
ax.plot(brownnoise_wavdata[:brownnoise_fs*30])
ax.set_title("Brownnoise based on NHK Radio English Signal")
ax.set_xlabel("time [pt]")
ax.set_ylabel("amplitude")
ax.set_ylim([-0.04, 0.04])

ax = fig.add_subplot(414)
ax.plot(tpdfnoise_wavdata[:tpdfnoise_fs*30])
ax.set_title("Tpdfnoise based on NHK Radio English Signal")
ax.set_xlabel("time [pt]")
ax.set_ylabel("amplitude")
ax.set_ylim([-0.04, 0.04])

fig.set_tight_layout(True)

plt.savefig('./Output/SoundFile_Noise_Signal_Plot.jpg')
plt.show()
```
